```python
from fastquant import get_crypto_data # extract crypto data from package
crypto = get_crypto_data("BTC/USDT", "2021-05-01", "2021-07-17") # can any date
crypto.head()
```

```python
from matplotlib import pyplot as plt # past data, but not predictive 
crypto.close.plot(figsize=(12, 7))
plt.title("Daily Closing Price of BTC/USDT", fontsize=20)
```

```python
type (crypto) # Has to be pandas to fit the pbprophet model we're using
```

```python
crypto.dtypes
```

```python
#converting data types from index

crypto ['Year'] = crypto.index
crypto.reset_index(level = 0, inplace = True)
```

```python
crypto.dtypes
```

```python
pip install fbprophet # Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data. Prophet is robust to missing data and shifts in the trend, and typically handles outliers well.
```

```python
from fbprophet import Prophet
import time
import datetime
```

```python
# need to rename columns in order to fit the data to have ds & y from the table
crypto = crypto.rename(columns= {'dt':'ds','close': 'y'})
```

```python
# Creating & Fitting the Model to data
model = Prophet(interval_width=0.95)
model.fit(crypto)
```

```python
future_dates = model.make_future_dataframe(periods=1, freq='MS')
future_dates.tail()
```

```python
#Forcasting Future Dates
forecast = model.predict(future_dates)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

#Yhat: predictive value
#Yhat_lower/upper: uncertainty interval
```

```python
 model.plot_components(forecast)
```

```python
import numpy as np
import pandas as pd
import requests
import json
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from datetime import datetime
import matplotlib.ticker as ticker


from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()


#switch between cn and us for different date formats
#change limit for number of days before today
FG_url = "https://api.alternative.me/fng/?limit=60&format=json&date_format=cn"
FG_JSONContent = requests.get(FG_url).json()
FG_content = json.dumps(FG_JSONContent, indent = 4, sort_keys=True)
FG_json_dict = json.loads(FG_content)
FG_json_list = FG_json_dict['data']

FG_dates = []
for item in FG_json_list:
    date = (item['timestamp'])
    FG_dates.append(date)
  
FG_values = []
for item in FG_json_list:
    item_int = int(item['value'])
    FG_values.append(item_int)

fig, ax = plt.subplots(figsize=(12, 7))
ax.set(ylabel="Value")
plt.title("Daily Crypto Fear and Greed Index", fontsize=20)
converted_dates = matplotlib.dates.datestr2num(FG_dates)
ax.xaxis_date()
ax.xaxis.set_minor_locator(ticker.LinearLocator(numticks=30))
ax.xaxis.set_minor_formatter(DateFormatter("%d"))
ax.xaxis.set_major_locator(ticker.IndexLocator(base=30, offset=11.7))
ax.xaxis.set_major_formatter(DateFormatter("\n\n%b\n%Y"))
plt.plot(converted_dates, FG_values, color='pink')

plt.show()
```
