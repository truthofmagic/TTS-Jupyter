# Basic Visualization

## Histograms

```r
x <- sample(1:100, 1000, replace=TRUE)
hist(x)
```

## Basic Plotting

```r
x <- rnorm(100, mean=0, sd=1)
y <- x + rnorm(100, mean=0, sd=1)
plot(x,y)
```

```r
x.s <- sort(x)
y.s <- sort(y)
plot(x.s, y.s); lines(x.s, y.s); abline(a=0, b=1, col="red")
```
