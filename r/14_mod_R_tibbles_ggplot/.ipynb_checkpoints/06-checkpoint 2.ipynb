## geom versus stat

You can usually use ```geom()``` or ```stat()``` interchangeably.  Every geom has a default stat; and vice versa.

````r
p1 <- ggplot(data=diamonds) + geom_bar(aes(x=cut))
````

````r
p2 <- ggplot(data=diamonds) + stat_count(aes(x=cut))
````

````r
p1
````

````r
p2
````
